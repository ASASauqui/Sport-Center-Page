# Encabezado del Excel

```python
# Consulta
r = requests.get("https://www.thesportsdb.com/api/v1/json/1/lookup_all_teams.php?id="+str(4328))
data = r.json()


data1 = list(data['teams'][0])
data1.append('idCountry')
# print(data1)
data2 = []
idsCheck = [data['teams'][0]['idTeam']]

for k in data['teams'][0].values():
    if(isinstance(k,str) == True and k != ''):
        data2.append(str(k))
    else:
        data2.append('null')
data2[52] = data2[52].replace('\t', '')
data2[52] = data2[52].replace('\n', '')
data2[52] = data2[52].replace('\"', '')


for k in range(0,len(paisesLista)-1):
    if(paisesLista[k] == data2[52]):
        data2.append(k)
        break
    

with open("my_data.csv", "a", newline='', encoding="utf-16") as csvfile:
    wr = csv.writer(csvfile,delimiter='\t', dialect = 'excel')
    wr.writerow(data1)
    wr.writerow(list(data2))
```

# Datos del Excel

```python
print('INICIO')
for i in ids:
    r = requests.get("https://www.thesportsdb.com/api/v1/json/1/lookup_all_teams.php?id="+str(i))
    data = r.json()
    if(isinstance(data['teams'],list)):            
        noData = len(data['teams'])
        for j in range(0, noData-1):
            equipoID = data['teams'][j]['idTeam']
            idsCheckBool = True
            if(equipoID in idsCheck):
                idsCheckBool = False
            if(idsCheckBool == True):
                idsCheck.append(equipoID)
                
            if(idsCheckBool == True):
                data1 = list(data['teams'][j].values())
                data2 = []
                for k in data1:
                    if(isinstance(k,str) == True and k != ''):
                        data2.append(str(k))
                    else:
                        data2.append('null')
                        
                data2[52] = data2[52].replace('\t', '')
                data2[52] = data2[52].replace('\n', '')
                data2[52] = data2[52].replace('\"', '')
                        
                if(data2[52] == 'null'):
                    data2.append('null')
                else:
                    for k in range(0,len(paisesLista)-1):
                        if(paisesLista[k] == data2[52]):
                            data2.append(k)
                            break

                with open("my_data.csv", "a", newline='', encoding="utf-16") as csvfile:
                    wr = csv.writer(csvfile,delimiter='\t', dialect = 'excel')
                    wr.writerow(data2)
print('FIN')
```
