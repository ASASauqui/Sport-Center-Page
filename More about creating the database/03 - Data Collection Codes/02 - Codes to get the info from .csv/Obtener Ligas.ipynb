```python
with open('Ligas.csv','r', newline ='', encoding="utf-16") as csvfile:
    r = csv.reader(csvfile, delimiter='\t')
    data = list(r)
    data[0].append('idPais')
    data[0].append('idDeporte')
    
    print(data[0])
    
csvfile.close()
    
    
with open("nuevasLigas.csv", "a", newline='', encoding="utf-16") as csvfile1:
    wr = csv.writer(csvfile1,  delimiter='\t', dialect = 'excel')
    wr.writerow(data[0])
    
csvfile1.close() 

data.pop(0)
```

# Datos del Excel

```python
# Paises
with open('Libro1Paises.csv','r', newline ='') as csvfile:
    r = csv.reader(csvfile, delimiter=',')
    paises = list(r)
    paises.pop(0)
    
    for row in data:
        row[4] = row[4].replace('\t', '')
        row[4] = row[4].replace('\n', '')
        row[4] = row[4].replace('\"', '')
    
        
    for row in data:
        for i in paises:
            if row[4] == i[1]:
                row.append(i[0])
                break
    
csvfile.close()








# Deportes
with open('Libro2Deportes.csv','r', newline ='') as csvfile:
    r = csv.reader(csvfile, delimiter=',')
    deportes = list(r)
    deportes.pop(0)
    
    for row in data:
        for i in deportes:
            if row[2] == i[1]:
                row.append(i[0])
                break
                
    for row in data:
        print(len(row))
        
csvfile.close()

    
    
    

    

    
    
    
with open("nuevasLigas.csv", "a", newline='', encoding="utf-16") as csvfile1:
    wr = csv.writer(csvfile1,  delimiter='\t', dialect = 'excel')
    
    for row in data:
         wr.writerow(row)

csvfile1.close() 
```
